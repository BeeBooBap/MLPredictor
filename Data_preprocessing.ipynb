```python
# Data Pre-processing 
```

```python
# filtering possible categorical values

cat_df = df.select_dtypes(include=['object']).copy()

cat_df.head()
```

```python
print(cat_df.shape)
```

```python
# count total number of missing values

print(cat_df.isnull().sum())
```

```python
# mode imputation for null values

for col in cat_df.columns:
  cat_df = cat_df.fillna(cat_df[col].value_counts().index[0])
```

```python
print(cat_df.isnull().sum())
```

```python
# casting categorical variables to category dtype

cat_df_lc = cat_df.copy()

for col in cat_df.columns:
  cat_df_lc[col] = cat_df_lc[col].astype('category')

cat_df_lc.dtypes
```

```python
# label encoding to convert categorical variables

for name in cat_df_lc.columns:
  cat_df_lc[name] = cat_df_lc[name].cat.codes

cat_df_lc.head()
```

```python
# dropping previous categorical variables from original df

old_df = df.copy()

for col in cat_df.columns:
  del old_df[col]

print (old_df)
```

```python
print(old_df.isnull().sum().sort_values(ascending=False))
```

```python
# dropping values with high null value count that are not relevant 

old_df.drop("adminActionState", axis=1, inplace=True)
old_df.drop("authorityDecision2", axis=1, inplace=True)
old_df.drop("petitionerState", axis=1, inplace=True)
old_df.drop("caseSourceState", axis=1, inplace=True)
old_df.drop("caseOriginState", axis=1, inplace=True)
old_df.drop("respondentState", axis=1, inplace=True)
old_df.drop("adminAction", axis=1, inplace=True)
```

```python
print(old_df.shape)
print(old_df.isnull().sum().sort_values(ascending=False))
```

```python
# mean imputation for null values

updated_df = old_df.copy()

for col in updated_df.columns:
  updated_df = updated_df.fillna(updated_df[col].mean())

print(updated_df.isnull().sum().sort_values(ascending=False))
```

```python
# combining with new label encoded variables

numeric_df = updated_df.join(cat_df_lc)

print(numeric_df)
print(numeric_df.shape)
print(numeric_df.dtypes)
```
